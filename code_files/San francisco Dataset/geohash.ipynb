```python
from shapely import from_wkt

import pandas as pd

import geohash
```

```python
hexes_df = pd.read_csv('csv_files/hex_data')
hexes_df['geometry'] = hexes_df['geometry'].apply(from_wkt)
```

```python
hex_geohash = []
for hex in hexes_df.values:
    #I've choosen precision 6, it's around 1220 meter radius
    hex_geohash.append([hex[0],geohash.encode(hex[1].centroid.y,hex[1].centroid.x,6)])
hex_geohash_df = pd.DataFrame(hex_geohash,columns=['hex_id','hex_geohash'])
```

```python
hex_geohash_df.to_csv('csv_files/hex_geohash',index=False)
```
